```python
#Part 3: Data Visualization
#12. Plot the distribution of median_house_value with a histogram.
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['MedHouseVal'], bins=10,edgecolor='black')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.title('Distribution of Median House Value')
plt.show()
```

```python
# Create a scatter plot of longitude vs latitude, colored by median_house_value.
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['MedHouseVal'], cmap='viridis', edgecolors='k')
plt.colorbar(scatter, label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot:Longitude vs Latitude colored by Median House Value')
plt.show()
```

```python
#14. Plot a boxplot of median_house_value grouped by the new income categories.
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x='IncomeCategory', y='MedHouseVal', data=df)

plt.xlabel('Income Category')
plt.ylabel('Median House Value')
plt.title('Boxplot of Median House Value Grouped by Income Category')
plt.show()
```
